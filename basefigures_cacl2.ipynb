FFT and iFFT Plot

```python
# try cleaning one signal with iFFT
from scipy.fft import rfft, irfft, rfftfreq
fs = 30000.0

dt = 1.0 / fs
segment = crcpost1_seg
N = len(segment)
f = rfftfreq(N, dt)

F = rfft(segment)


plt.figure(figsize=(10,8))
plt.vlines(f[:N//2], 0,  2.0/N * np.abs(F[0:N//2]), label='Post-CaCl2 Treatment 1', color='blue')
plt.xlim(-3, 1500) # limit x-axis to 1500 Hz
plt.axvline(x=500, linestyle='--', color='black')
#plt.ylim(0,5)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (µV)')
plt.legend()
```

```python
#plot iFFT cleaned signal
from scipy.fft import ifft
F_cleaned = F.copy()
step=10
# Zero out frequencies above 400 Hz
F_cleaned[np.abs(f) > 500] = 0         
signal_cleaned = irfft(F_cleaned)
plt.figure(figsize=(35,12))
plt.plot(signal_cleaned.real[start:end], label='iFFT Cleaned Signal', color='purple')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude (µV)')
plt.title('iFFT Cleaned Signal (Low Pass Filter at 500 Hz)')
plt.legend()
```
