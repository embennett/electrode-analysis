Read in Post-treatment Media Recording

```python
# read in post-treatment media reading
base_name = all_xdat_datasource_names[4] # first base name
print("Will work with file %s" % base_name)
file_path_mediaPost = str(Path(target_dir, base_name))
time_range = afr.get_allego_xdat_time_range(file_path_mediaPost)
time_range
```

```python
time_start = 0
time_mid = time_range[1]/2
time_end = time_range[1]

pri_signals_mediaPost_seg1, timestamps_mediaPost_seg1, time_samples_mediaPost_seg1 = afr.read_allego_xdat_pri_signals(file_path_mediaPost, 
                                                                         time_start, time_mid)
pri_signals_mediaPost_seg2, timestamps_mediaPost_seg2, time_samples_mediaPost_seg2 = afr.read_allego_xdat_pri_signals(file_path_mediaPost, 
                                                                         time_mid, time_end)

print(pri_signals_mediaPost_seg1.shape)
```

```python
#this is just for one cite to check how the graph looks

id=5

start = int(0)
end   = int(5e6) # fix to be the full index range

#segment 1 times
times_mediaPre_seg1 = time_samples_mediaPre_seg1[start:end]
times_mediaPost_seg1 = time_samples_mediaPost_seg1[start:end]
times_crcCtrl_seg1 = time_samples_crcCtrl_seg1[start:end]
times_crcPre_seg1 = time_samples_crcPre_seg1[start:end]
times_crcpost1_seg1 = time_samples_crcPost1_seg1[start:end]
times_crcpost2_seg1 = time_samples_crcPost2_seg1[start:end]

#segment 2 times
times_mediaPre_seg2 = time_samples_mediaPre_seg2[start:end]
times_mediaPost_seg2 = time_samples_mediaPost_seg2[start:end]
times_crcCtrl_seg2 = time_samples_crcCtrl_seg2[start:end]
times_crcPre_seg2 = time_samples_crcPre_seg2[start:end]
times_crcpost1_seg2 = time_samples_crcPost1_seg2[start:end]
times_crcpost2_seg2 = time_samples_crcPost2_seg2[start:end]

#segment 1 signals
mediaPre_seg1 = pri_signals_mediaPre_seg1[id][start:end]
mediaPost_seg1  = pri_signals_mediaPost_seg1[id][start:end]
crcCtrl_seg1 = pri_signals_crcCtrl_seg1[id][start:end]
crcpre_seg1 = pri_signals_crcPre_seg1[id][start:end]
crcpost1_seg1 = pri_signals_crcPost1_seg1[id][start:end]
crcpost2_seg1 = pri_signals_crcPost2_seg1[id][start:end]

#segment 2 signals

mediaPre_seg2 = pri_signals_mediaPre_seg2[id][start:end]
mediaPost_seg2  = pri_signals_mediaPost_seg2[id][start:end]
crcCtrl_seg2 = pri_signals_crcCtrl_seg2[id][start:end]
crcpre_seg2 = pri_signals_crcPre_seg2[id][start:end]
crcpost1_seg2 = pri_signals_crcPost1_seg2[id][start:end]
crcpost2_seg2 = pri_signals_crcPost2_seg2[id][start:end]
```

```python
# plot dose one and two next to each other
plt.figure(figsize=(16, 10))
plt.subplot(211)
plt.plot(times_crcpost1_seg1, crcpost1_seg1, alpha=0.7, label='CRC Post-KCl Treatment 1', color='orange')
plt.plot(times_crcpost1_seg2, crcpost1_seg2, alpha=0.7, color='orange')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (µV)')
plt.title('Electrode ID %d Response to KCl Treatment - Dose 1' % id)
#plt.xlim(0, 38)
#plt.ylim(-6000, 6000)
plt.legend()

#does 2
plt.subplot(212)
plt.plot(times_crcpost2_seg1, crcpost2_seg1, alpha=0.7, label='CRC Post-KCl Treatment 2', color='blue')
plt.plot(times_crcpost2_seg2, crcpost2_seg2, alpha=0.7, color='blue')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude (µV)')
plt.title('Electrode ID %d Response to KCl Treatment - Dose 2' % id)
#plt.xlim(0, 38)
#plt.ylim(-6000, 6000)
plt.legend()
```

```python
# plot id 1-10 for each dose
plt.figure(figsize=(16, 50))
for id in range(1, 11):
    plt.subplot(10, 2, id*2-1)
    plt.plot(times_crcpost1_seg1, pri_signals_crcPost1_seg1[id][start:end], alpha=0.7, label='CRC Post-KCl Treatment 1', color='orange')
    plt.plot(times_crcpost1_seg2, pri_signals_crcPost1_seg2[id][start:end], alpha=0.7, color='orange')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (µV)')
    plt.title('Electrode ID %d Response to KCl Treatment - Dose 1' % id)
    #plt.xlim(0, 38)
    #plt.ylim(-6000, 6000)
    plt.legend()

for id in range(1, 11):
    plt.subplot(10, 2, id*2)
    plt.plot(times_crcpost2_seg1, pri_signals_crcPost2_seg1[id][start:end], alpha=0.7, label='CRC Post-KCl Treatment 2', color='blue')
    plt.plot(times_crcpost2_seg2, pri_signals_crcPost2_seg2[id][start:end], alpha=0.7, color='blue')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (µV)')
    plt.title('Electrode ID %d Response to KCl Treatment - Dose 2' % id)
    #plt.xlim(0, 38)
    #plt.ylim(-6000, 6000)
    plt.legend()
```
