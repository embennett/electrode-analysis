```python
# create all time segments
start = int(0)
end   = int(5e6) # fix to be the full index range

#segment 1 times
#times_mediaPre_seg1 = time_samples_mediaPre_seg1[start:end]
times_mediaPost_seg1 = time_samples_mediaPost_seg1[start:end]
times_crcCtrl_seg1 = time_samples_crcCtrl_seg1[start:end]
times_crcPre_seg1 = time_samples_crcPre_seg1[start:end]
times_crcpost1_seg1 = time_samples_crcPost1_seg1[start:end]
times_crcpost2_seg1 = time_samples_crcPost2_seg1[start:end]
times_crcpost3_seg1 = time_samples_crcPost3_seg1[start:end]

#segment 2 times
#times_mediaPre_seg2 = time_samples_mediaPre_seg2[start:end]
times_mediaPost_seg2 = time_samples_mediaPost_seg2[start:end]
times_crcCtrl_seg2 = time_samples_crcCtrl_seg2[start:end]
times_crcPre_seg2 = time_samples_crcPre_seg2[start:end]
times_crcpost1_seg2 = time_samples_crcPost1_seg2[start:end]
times_crcpost2_seg2 = time_samples_crcPost2_seg2[start:end]
times_crcpost3_seg2 = time_samples_crcPost3_seg2[start:end]
```

```python
#figure saving directory
main_dir = os.getcwd()                   # current working directory
figures_dir = os.path.join(main_dir, "Figures/KCl_Exp2/Org1")
os.makedirs(figures_dir, exist_ok=True) # create folder if it doesn't exist
```

```python
# dont need the plot past 50 seconds
# plot id 1-10 for each dose
for id in range(1, 11):
    plt.figure(figsize=(16, 6))
    plt.subplot(131)
    plt.plot(times_crcpost1_seg1, pri_signals_crcPost1_seg1[id][start:end], alpha=0.7, label='CRC Post-KCl Dose 1', color='orange')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (µV)')
    plt.title('Electrode ID %d Response to KCl Treatment - Dose 1' % id)
    plt.xlim(0, 50)
    plt.ylim(-8000, 8000)
    plt.legend()

    plt.subplot(132)
    plt.plot(times_crcpost2_seg1, pri_signals_crcPost2_seg1[id][start:end], alpha=0.7, label='CRC Post-KCl Dose 2', color='blue')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (µV)')
    plt.title('Electrode ID %d Response to KCl Treatment - Dose 2' % id)
    plt.xlim(0, 50)
    plt.ylim(-8000, 8000)
    plt.legend()

    plt.subplot(133)
    plt.plot(times_crcpost3_seg1, pri_signals_crcPost3_seg1[id][start:end], alpha=0.7, label='CRC Post-KCl Dose 3', color='darkpink')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude (µV)')
    plt.title('Electrode ID %d Response to KCl Treatment - Dose 3' % id)
    plt.xlim(0, 50)
    plt.ylim(-8000, 8000)
    plt.legend()

    save_path = os.path.join(figures_dir, f"kclexp1_comparison_channel{id}.png")
    plt.savefig(save_path, dpi=300)

```
